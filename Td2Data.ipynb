Reconstruction de la base de données

Combien de service providers ont été déclarés dans les offshore leaks et quelle est la proportion de ces service providers vis-à-vis de des officers ?

```python
import pandas as pd
df=pd.read_csv("relationships.csv", low_memory=False)
dr = pd.read_csv("nodes-officers.csv",low_memory=False)
```

```python
value = df[df["_type"] == "officer_of"]

```

```python
m =pd.merge(dr,value,how="inner",left_on = "_id",right_on ="_start")
m
```

```python
dv = pd.read_csv("nodes-entities.csv",low_memory=False)
```

```python
merge = m.merge(dv,left_on="_end",right_on ="_id").groupby(by="service_provider").count()
```
